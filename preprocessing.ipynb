```python
import pandas as pd

data_path = './hospital_data.csv'

df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
df.columns
```

```python
## null이 있는 행 삭제

df = df.dropna()

## check null

df.isnull().sum()
```

```python
df.count()
```

```python
## '경도' '위도' column 삭제

df = df.drop(['경도', '위도'], axis=1)

df.head()
```

```python
df = df.drop(['Unnamed: 10'], axis=1)

df.head()
```

```python
from pyproj import Transformer

transformer = Transformer.from_crs("EPSG:5174", "EPSG:4326", always_xy=True)

def tm_to_latlon(x, y):
    lon, lat = transformer.transform(x, y)
    return lat, lon

# 4. 위도/경도 열 추가
df[['위도', '경도']] = df.apply(lambda row: tm_to_latlon(row['좌표정보(x)'], row['좌표정보(y)']), axis=1, result_type='expand')

# 5. 변환 결과 출력 및 저장 (필요 시)
print(df.head())
```

```python
print(df['위도'].count())
print(df['경도'].count())
print(len(df))
```

```python
# save to csv 

df.to_csv('./hospital_data_processed.csv', index=False)
```
